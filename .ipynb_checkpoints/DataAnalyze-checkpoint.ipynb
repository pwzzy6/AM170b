```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score
import warnings 
warnings.filterwarnings('ignore')
```

```python
ar = pd.read_csv('AirQuality.csv',sep=';')
ar
```

```python
# getting the numerical estimates of all the numerical column
ar.describe()
```

```python
#Dropping CO(GT) and Unnamed columns
ar.drop(['CO(GT)','Unnamed: 15','Unnamed: 16'],axis = 1,inplace = True)
```

```python
#Formatting some object columns from strings to floats

ar.replace(to_replace=',',value='.',regex=True,inplace=True) 

for i in 'C6H6(GT) T RH AH'.split():
    ar[i] = pd.to_numeric(ar[i],errors='coerce') 
```

```python
#Replacing null data from -200 to NaN for posterior treatment

ar.replace(to_replace=-200,value=np.nan,inplace=True)

ar.info()
```

```python
#Formatting Date and Time to datetime type

ar['Date'] = pd.to_datetime(ar['Date'],dayfirst=True) 

ar['Time'] = pd.to_datetime(ar['Time'],format= '%H.%M.%S' ).dt.time

ar.head()
```

```python
#Removing NMHC(GT) sensor due to amount of null values

ar.drop('NMHC(GT)', axis=1, inplace=True) 

ar.info()
```

```python
ar.drop(['CO(GT)']  ,axis=1, inplace=True, errors = 'ignore')
ar.drop(['NOx(GT)'] ,axis=1, inplace=True, errors = 'ignore')
ar.drop(['C6H6(GT)'],axis=1, inplace=True, errors = 'ignore')
ar.drop(['NO2(GT)'] ,axis=1, inplace=True, errors = 'ignore')
ar.head()
```

```python
ar = ar.dropna()
percent_NaN = []
columns = ar.columns
for col in columns:
    pNaN =  (ar[col].isna().sum()/ar.shape[0]) * 100 #sum NaN instances in each column. Divide by total rows
    percent_NaN.append(pNaN)
nan_percent_df = pd.DataFrame(percent_NaN,
                              index=columns,
                              columns=['%_NaN_in_Column']).sort_values('%_NaN_in_Column',ascending = False)
nan_percent_df
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('AirQuality.csv',sep=";", decimal=",",header=0)

df.head()
df.drop(columns=['Unnamed: 15', 'Unnamed: 16'],inplace=True)
df.dropna(inplace=True)
df.replace(to_replace=-200,value=np.nan,inplace=True)

df.to_csv("uncleaned data.csv", index=False)
# df.drop(columns=['NMHC(GT)'],inplace=True)
# df.isna().sum()

# df.drop(columns=['C6H6(GT)'],inplace=True)
# df.drop(columns=['NOx(GT)'],inplace=True)
# df.drop(columns=['NO2(GT)'],inplace=True)
# df.drop(columns=['CO(GT)'],inplace=True)
```

```python
col = ['PT08.S1(CO)','PT08.S2(NMHC)', 'PT08.S3(NOx)', 'PT08.S4(NO2)','PT08.S5(O3)', 'T', 'RH', 'AH']
df = df[col]
df[col].dtypes
df.describe()
```

```python
for i in col:
    df[i] = df[i].fillna(df[i].mean())
plt.figure(figsize=(5,5))
sns.boxplot(data=df)
plt.xticks(rotation='vertical')
plt.show()
```

```python
# getting the quartile one and quartile 3 values of each column
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
# finally calculating the interquartile range IQR
IQR = Q3 - Q1
((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()
```

```python
mask = (df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))
mask
```

```python
# now replacing all the outliers using the median of that particular column
for i in mask.columns:
    df[i].astype('float')
    temp = df[i].median()
    df.loc[mask[i], i] = temp
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(data=df)
plt.xticks(rotation='vertical')
plt.show()
```

```python
df.corr()
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load your data
data = pd.read_csv('uncleaned data.csv')

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data.select_dtypes(include=[float, int]))

# Apply PCA to reduce to 2 principal components
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

# Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=4)
hierarchical_clusters = hierarchical.fit_predict(pca_data)

# DBSCAN Clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_clusters = dbscan.fit_predict(pca_data)

# Gaussian Mixture Model Clustering
gmm = GaussianMixture(n_components=4, random_state=42)
gmm_clusters = gmm.fit_predict(pca_data)

# Create DataFrames for each clustering method
clustered_data_hierarchical = pd.DataFrame(pca_data, columns=['PCA1', 'PCA2'])
clustered_data_hierarchical['Cluster'] = hierarchical_clusters

clustered_data_dbscan = pd.DataFrame(pca_data, columns=['PCA1', 'PCA2'])
clustered_data_dbscan['Cluster'] = dbscan_clusters

clustered_data_gmm = pd.DataFrame(pca_data, columns=['PCA1', 'PCA2'])
clustered_data_gmm['Cluster'] = gmm_clusters

# Save the clustered data to new CSV files
# clustered_data_hierarchical.to_csv('/path/to/save/clustered_data_hierarchical.csv', index=False)
# clustered_data_dbscan.to_csv('/path/to/save/clustered_data_dbscan.csv', index=False)
# clustered_data_gmm.to_csv('/path/to/save/clustered_data_gmm.csv', index=False)

# Plot the results
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
scatter = plt.scatter(clustered_data_hierarchical['PCA1'], clustered_data_hierarchical['PCA2'], c=clustered_data_hierarchical['Cluster'], cmap='viridis', marker='o')
plt.title('Hierarchical Clustering Results')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(*scatter.legend_elements(), title="Clusters")
plt.grid(True)

plt.subplot(1, 3, 2)
scatter = plt.scatter(clustered_data_dbscan['PCA1'], clustered_data_dbscan['PCA2'], c=clustered_data_dbscan['Cluster'], cmap='viridis', marker='o')
plt.title('DBSCAN Clustering Results')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(*scatter.legend_elements(), title="Clusters")
plt.grid(True)

plt.subplot(1, 3, 3)
scatter = plt.scatter(clustered_data_gmm['PCA1'], clustered_data_gmm['PCA2'], c=clustered_data_gmm['Cluster'], cmap='viridis', marker='o')
plt.title('GMM Clustering Results')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(*scatter.legend_elements(), title="Clusters")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Load your data
data = pd.read_csv('cleaned data')

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data.select_dtypes(include=[float, int]))

# Apply PCA to reduce to 2 principal components
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

# Perform Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=2)
hierarchical_clusters = hierarchical.fit_predict(pca_data)

# Create DataFrame for hierarchical clustering results
clustered_data_hierarchical = pd.DataFrame(pca_data, columns=['PCA1', 'PCA2'])
clustered_data_hierarchical['Cluster'] = hierarchical_clusters

# Save the clustered data to a new CSV file
clustered_file_path_hierarchical = 'cleanDataClustered.csv'
clustered_data_hierarchical.to_csv(clustered_file_path_hierarchical, index=False)

# Plot the results with different colors for each cluster and labeled PCA components
plt.figure(figsize=(10, 6))
scatter = plt.scatter(clustered_data_hierarchical['PCA1'], clustered_data_hierarchical['PCA2'], c=clustered_data_hierarchical['Cluster'], cmap='viridis', marker='o')

# Add titles and labels
plt.title('Hierarchical Clustering Results')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')

# Add a legend
plt.legend(*scatter.legend_elements(), title="Clusters")

# Show the plot
plt.grid(True)
plt.show()
```
