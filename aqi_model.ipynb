```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

# Load the data
file_path = 'combined_data.csv'
data = pd.read_csv(file_path)

data['AQI_NO2'] = data['AQI_NO2'] / 3
data['AQI_O3'] = data['AQI_O3'] / 2
data['AQI_PM25'] = data['AQI_PM25'] / 3.2
data['AQI_SO2'] = data['AQI_SO2'] * 1.5
data['AQI_PM10'] = data['AQI_PM10'] / 2.8
data['PT08.S1(CO)'] = data['PT08.S1(CO)'] * 0.01
data['PT08.S2(NMHC)'] = data['PT08.S2(NMHC)'] * 0.01
data['PT08.S3(NOx)'] = data['PT08.S3(NOx)'] * 0.01
data['PT08.S4(NO2)'] = data['PT08.S4(NO2)'] * 0.01
data['PT08.S5(O3)'] = data['PT08.S5(O3)'] * 0.01

data['Mean_Feature'] = data[['PT08.S1(CO)', 'PT08.S2(NMHC)', 'PT08.S3(NOx)', 'PT08.S4(NO2)', 'PT08.S5(O3)', 'AQI_CO', 'AQI_NO2', 'AQI_O3', 'AQI_PM10',	'AQI_PM25',	'AQI_SO2','Overall_AQI']].mean(axis=1)

data
```

```python
# Ensure the Start column is in datetime format
data['Start'] = pd.to_datetime(data['Start'])

# Sort the data by DateTime
data.sort_values(by='Start', inplace=True)

# Set the Start column as the index
data.set_index('Start', inplace=True)

# Create lagged feature for Mean_Feature
data['Mean_lagged'] = data['Mean_Feature'].shift(1)

# Drop rows with NaN values
data.dropna(inplace=True)

# Define input feature (lagged Mean) and output target (current Mean)
X = data['Mean_lagged'].values.reshape(-1, 1)
y = data['Mean_Feature'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)

# Train a simple linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions for the entire dataset
y_pred = model.predict(X)

# Create a DataFrame for the results
results = pd.DataFrame({
    'Start': data.index,
    'Actual_Mean': y,
    'Predicted_Mean': y_pred
})

# Plot the actual data and predictions
plt.figure(figsize=(14, 7))
plt.plot(results['Start'], results['Actual_Mean'], label='Actual AQI+')
plt.plot(results['Start'], results['Predicted_Mean'], label='Predicted AQI+', linestyle='--')
plt.xlabel('DateTime')
plt.ylabel('AQI+ Levels')
plt.title('Actual vs Predicted AQI+ Over Time')
plt.legend()
plt.show()
```
